1. Check if the array is sorted using recursion. Draw recurrence tree.

```python
def sortedArrayWithRecursion(l1):
    if len(l1) in [0, 1]:
        return True

    restSortedArray = sortedArrayWithRecursion(l1[1:])
    if restSortedArray == False:
        return False

    if l1[0] < l1[1]:
        return True
    else:
        return False
```

```python
sortedArrayWithRecursion([2, 5, 7, 9, 1])
```

```python
def checkSortedArray(ls: list) -> bool:
    if len(ls) in [0, 1]:
        return True

    if ls[0] > ls[1]:
        return False

    return checkSortedArray(ls[1:])
```

```python
checkSortedArray([1, 2, 3, 4, 5, 6,])
```
