1. Find 2 to the power of n for a given n, using recursion tree. Also state recursion function and make recursion tree.

````python
def powerOf2(n: int) -> int:
    if n == 1:
        return 2
    else:
        return 2 * powerOf2(n-1)


def powerOf2_new(n: int) -> int:
    if n != 1:
        return 2 * powerOf2_new(n-1)
    return 2
````

````python
print(powerOf2(5),
powerOf2_new(4))
````

2. Find the number of digits in a given number
```
234 -> 3
51234 -> 5
```
draw recusrion tree and recursion formula

````python
def find_digits(num: int) -> int:
    if num == 0:
        return 0
    elif num < 10:
        return 1
    else:
        return 1 + find_digits(num//10)
````

````python
find_digits(122)
````

3. Fibonicci Series

````python
def fibonicci_(num: int) -> int:
    if num == 0:
        return 1
    elif num == 1:
        return 1
    else:
        return fibonicci_(num-2) + fibonicci_(num-1)
````

````python
fibonicci_(8)
````
