1. Find 2 to the power of n for a given n, using recursion tree. Also state recursion function and make recursion tree.

````python
def powerOf2(n: int) -> int:
    if n == 1:
        return 2
    else:
        return 2 * powerOf2(n-1)


def powerOf2_new(n: int) -> int:
    if n != 1:
        return 2 * powerOf2_new(n-1)
    return 2
````

````python
print(powerOf2(5),
powerOf2_new(4))
````

2. Find the number of digits in a given number
```
234 -> 3
51234 -> 5
```
draw recusrion tree and recursion formula

````python
def find_digits(num: int) -> int:
    if num == 0:
        return 0
    elif num < 10:
        return 1
    else:
        return 1 + find_digits(num//10)
````

````python
find_digits(122)
````

3. Fibonicci Series

````python
def fibonicci_(num: int) -> int:
    if num in [0, 1]:
        return 1
    else:
        return fibonicci_(num-2) + fibonicci_(num-1)
````

````python
fibonicci_(5)
````

````python
# with all function's call

callOfFunction = 1


def fibonicci_new(num: int) -> int:
    global callOfFunction
    print("Function's Call", callOfFunction, "---->", num)
    callOfFunction += 1
    if num in [0, 1]:
        return 1
    else:
        return fibonicci_new(num - 1) + fibonicci_new(num - 2)
````

````python
fibonicci_new(5)
````

5. Sum all the nuber

````python
def sumAll(n):
    if n < 10:
        return n
    last = n % 10
    return last + sumAll(n//10)
````

````python
sumAll(102)
````

````python
def sumAll_new(n : int, ans: int)->int:
    if n == 0:
        return 0 + ans
    new = ans + n%10
    return sumAll_new(n//10, new)
````

````python
sumAll_new(234, 0)
````
