### Embeddings

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

# data_path = r"data\Practical Mobile Forensics.pdf"
data_path2 = r"C:\Users\D S Patwal\Downloads\Ai\Anthropic-enterprise-ebook-digital.pdf"

data = PyPDFLoader(data_path2).load()
```

```python
splitters = RecursiveCharacterTextSplitter(chunk_size=5000, chunk_overlap=1000)
data_chunks = splitters.split_documents(data)
```

```python
import os
from langchain_ollama.embeddings import OllamaEmbeddings
from dotenv import load_dotenv
load_dotenv()

nomic_embedding = OllamaEmbeddings(model=os.getenv("NOMIC_EMBEDDING"))
gemma_embedding = OllamaEmbeddings(model=os.getenv("GEMMA_EMBEDDING"))
mxbai_embedding = OllamaEmbeddings(model=os.getenv("MXBAI_EMBEDDING"))
```

```python
# Create FAISS vector store from documents using the chosen embedding
vector_store = FAISS.from_documents(data_chunks, nomic_embedding)
vector_store.save_local("faiss_index")
```

```python
new_data = r"G:\My Drive\BOOKS\Andrew Carnegie.pdf"


def upating_VDB(data: str):
    docs = PyPDFLoader(data).load()
    chunks = RecursiveCharacterTextSplitter(
        chunk_size=5000, chunk_overlap=1000).split_documents(docs)

    vector_store = FAISS.load_local(
        "faiss_index", embeddings=nomic_embedding, allow_dangerous_deserialization=True)

    # `add_documents` returns added ids — keep the same vector_store instance
    vector_store.add_documents(chunks)
    print(f"{data} saved")
    return vector_store.as_retriever()
```

```python
retriever = upating_VDB(new_data)
```
