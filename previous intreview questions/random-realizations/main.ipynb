- Suppose you have an function which can return Any dtype. You want to store them in a file if the output is str


```python
import random

function1_output = ["its a string", 200, True, 401]
result = random.choice(function1_output)

# answer
if type(result) == str:
    print(result)

# or use is keyword as == operator
if type(result) is str:
    print(result)
```
