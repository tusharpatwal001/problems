```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris, load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, r2_score
import matplotlib.pyplot as plt
```

```python
x, y = load_iris(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Standardization on the dataset
scaler = StandardScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Supervised models -> Classification models

```python
# Decision Tree Classifier -> Classifier type model
model1 = DecisionTreeClassifier()
model2 = DecisionTreeClassifier()

# fitting the models
model1.fit(x_train, y_train)
model2.fit(x_train_scaled, y_train)

# predictions of models
y_pred = model1.predict(x_test)
y_pred_scaled = model2.predict(x_test_scaled)

# show classification metrics
score = classification_report(y_test, y_pred)
score_scaled = classification_report(y_test, y_pred_scaled)

# Print all reports
print(f"Classification report for dataset without standardization: \n{score}")
print(f"Classification report for dataset with standardization: \n{score_scaled}")
```

```python
# Random Forest Classifier -> Classifier type model
model1 = RandomForestClassifier()
model2 = RandomForestClassifier()

# fitting the models
model1.fit(x_train, y_train)
model2.fit(x_train_scaled, y_train)

# predictions of models
y_pred = model1.predict(x_test)
y_pred_scaled = model2.predict(x_test_scaled)

# show classification metrics
score = classification_report(y_test, y_pred)
score_scaled = classification_report(y_test, y_pred_scaled)

# Print all reports
print(f"Classification report for dataset without standardization: \n{score}")
print(f"Classification report for dataset with standardization: \n{score_scaled}")
```

## Supervised models -> Regression models

```python
x, y = load_diabetes(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Standardization on the dataset
scaler = StandardScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

## Unsupervised models -> Clustering models

```python
x, y = load_iris(return_X_y=True)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(x)
```

```python
labels = kmeans.labels_  # Get the cluster labels assigned by K-Means
centers = kmeans.cluster_centers_  # Get the cluster centers

# Visualize the clusters (using the first two features for simplicity)
plt.figure(figsize=(8, 6))

# Plot data points, colored by their assigned cluster
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8)

# Plot cluster centers
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=200, color='red', label='Cluster Centers')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('K-Means Clustering of Iris Dataset (Sepal Length vs Sepal Width)')
plt.legend()
plt.grid(True)
plt.show()

# You can also visualize against the true labels for comparison
plt.figure(figsize=(8, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='viridis', s=50, alpha=0.8)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('True Species of Iris Dataset (Sepal Length vs Sepal Width)')
plt.grid(True)
plt.show()
```

## DBSCAN

```python
from sklearn.decomposition import PCA

iris = load_iris()
X = iris.data
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

```python
import numpy as np
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.25, min_samples=5)

dbscan_labels = dbscan.fit_predict(X_reduced)


unique_labels = np.unique(dbscan_labels)
n_clusters = len(unique_labels) - (1 if -1 in dbscan_labels else 0)  # Exclude noise from cluster count

plt.figure(figsize=(8, 6))
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, n_clusters + 1)]

for i, label in enumerate(unique_labels):
    if label == -1:  # Outliers
        plt.scatter(X_reduced[dbscan_labels == label][:, 0], X_reduced[dbscan_labels == label][:, 1],
                    s=20, c='k', label='Outliers')
    else:
        plt.scatter(X_reduced[dbscan_labels == label][:, 0], X_reduced[dbscan_labels == label][:, 1],
                    s=20, c=[colors[i]], label=f'Cluster {label}')

plt.title('DBSCAN Clustering on Iris Dataset (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
