```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from streamlit import scatter_chart
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='green')
plt.show()
```

## Basic Plots

```python
plt.plot(x, y)
```

```python
plt.bar(x, y)
```
